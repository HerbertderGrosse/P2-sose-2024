## Aufgabe 3: Detektorakzeptanz

 * Bestimmen Sie Rate aufgezeichneter Photonen für $_{\hphantom{0}55}^{137}\mathrm{Cs}$ bei fünf verschiedenen Abständen des Präparats von der Detektorstirnfläche.  
 * Schätzen Sie ab und begründen Sie, ob eine Korrektur auf den unterliegenden Untergrund notwendig ist. 
 * Schätzen Sie ab und begründen Sie, ob eine Korrektur des Detektors auf pile-up notwendig ist.

```python
hintergrund = Cs_1cm = np.genfromtxt("Mo 31\Hintergrund30min.hst")/6


Cs_1cm = np.genfromtxt("Mo 31\Caesium_5min_1cm.hst")-hintergrund
x = np.arange(len(Cs_1cm))


Cs_0cm = np.genfromtxt("Mo 31\Caesium_10min.hst")/2-hintergrund
Cs_2cm = np.genfromtxt("Mo 31\Caesium_5min_2cm.hst")-hintergrund

Cs_3cm = np.genfromtxt("Mo 31\Caesium_5min_3cm.hst")-hintergrund

Cs_4cm = np.genfromtxt("Mo 31\Caesium_5min_4cm.hst")-hintergrund

Cs_5cm = np.genfromtxt("Mo 31\Caesium_5min_5cm.hst")-hintergrund


plt.bar(x, Cs_0cm, width=1,label="kein Abstand")
plt.bar(x, Cs_1cm, width=1,label="1cm")
plt.bar(x, Cs_2cm, width=1,label="2cm")
plt.bar(x, Cs_3cm, width=1,label="3cm")
plt.bar(x, Cs_4cm, width=1,label="4cm")
plt.bar(x, Cs_5cm, width=1,label="5cm")
plt.xlabel("Channel")
plt.ylabel("Count")
plt.title("Darstellung von Cs-137 bei verschiedenen Abständen")
plt.legend()
plt.show()

plt.bar(x, Cs_0cm, width=1,label="kein Abstand")
plt.bar(x, Cs_1cm, width=1,label="1cm")
plt.bar(x, Cs_2cm, width=1,label="2cm")
plt.bar(x, Cs_3cm, width=1,label="3cm")
plt.bar(x, Cs_4cm, width=1,label="4cm")
plt.bar(x, Cs_5cm, width=1,label="5cm")
plt.xlabel("Channel")
plt.ylabel("Count")
plt.legend()
plt.semilogy()
plt.title("Darstellung von Cs-137 bei verschiedenen Abständen, logarithmisch")
plt.show()
```

```python
plt.bar(x, Cs_5cm+hintergrund, width=1,label="Spektrum+Hintergrund bei 5cm Abstand")
plt.bar(x, hintergrund*10, width=1,label="Hintergrundstrahlung 10-fach überhöht")
plt.bar(x, hintergrund, width=1,label="Hintergrundstrahlung")
plt.xlabel("Channel")
plt.ylabel("Count")
plt.legend()
plt.show()
```

```python
def dist(d,d1,n=1000000):
    return n/(d+d1)**2

Dist_Cont = kafe2.XYContainer([0,1,2,3,4,5],[sum(Cs_0cm),sum(Cs_1cm),sum(Cs_2cm),sum(Cs_3cm),sum(Cs_4cm),sum(Cs_5cm)])
Dist_Cont.add_error(axis='x',err_val=0.1)
Dist_Cont.add_error(axis='y',err_val=100000)
Dist_Cont.axis_labels=['Abstand [cm]','Anzahl registrierter Teilchen [n]']


Dist_Fit = kafe2.Fit(Dist_Cont,dist)
Dist_Fit.do_fit()

Dist_Plot = kafe2.Plot(Dist_Fit)
Dist_Plot.plot()
plt.show()
```

Es wurden sechs Messungen durchgeführt, dabei wurde der Abstand jeweils um $1\,cm$ erhöht. Eine Korrektur mit der Hintergrundstrahlung ist nicht notwendig, da die Ereignisrate sehr gering ist, sie wurde dennoch durchgeführt. Es ist zu sehen, dass der Einfluss der Hintergrundstrahlung um mindestens eine Größenordnung kleiner ist als die  Anzahl der Ereignisse von der Quelle. Eine Korrektur für pile-ups ist nicht notwendig, da sowieso wenige Teilchen ankommen und das Oszilloskop eine hohe zeitliche Auflösung hat. Das ist zu sehen in der logarithmischen Darstellung wo die Peaks der pile-ups um mindestens zwei Größenordnungen kleiner sind.

Als Modell für den Abfall wird $\frac{n}{(d+d_{\rm{Quelle}})^2}=n_{\rm{registriert}}$ verwendet.  Wir schätzen den Fehler auf den Abstand auf $1\,mm$ und auf die Teilchenmessung auf $100.000$. Für $n_{\rm{registriert}}$ wird die Gesammtanzahl an registrierten Ereignissen verwendet.

Das Modell des quadratischen Abfall scheint gut zu unseren Daten zu passen. Das $\frac{\chi^2}{ndf}$ ist mit 0.03041 klein, was an der relativ großen Freiheit  des Modells liegt. Die $\chi^2-\mathrm{probability}$ ist mit 0.998 gut.
