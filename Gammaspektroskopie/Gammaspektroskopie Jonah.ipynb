---

Die Messung wurde mit $1\,cm$ 


Eine Korrektur mit der Hintergrundstrahlung ist nicht notwendig, da die Ereignisrate sehr gering is. Eine Korrektur für pile-ups ist nicht notwendig, da sowieso wenige Teilchen ankommen und das Oszilloskop eine hohe zeitliche Auflösung hat. 

```python
hintergrund = Cs_1cm = np.genfromtxt("Mo 31\Hintergrund30min.hst")/6


Cs_1cm = np.genfromtxt("Mo 31\Caesium_5min_1cm.hst")-hintergrund
x = np.arange(len(Cs_1cm))


Cs_2cm = np.genfromtxt("Mo 31\Caesium_5min_2cm.hst")-hintergrund

Cs_3cm = np.genfromtxt("Mo 31\Caesium_5min_3cm.hst")-hintergrund

Cs_4cm = np.genfromtxt("Mo 31\Caesium_5min_4cm.hst")-hintergrund

Cs_5cm = np.genfromtxt("Mo 31\Caesium_5min_5cm.hst")-hintergrund

plt.bar(x, Cs_1cm, width=1)
plt.bar(x, Cs_2cm, width=1)
plt.bar(x, Cs_3cm, width=1)
plt.bar(x, Cs_4cm, width=1)
plt.bar(x, Cs_5cm, width=1)
plt.xlabel("Channel")
plt.ylabel("Count")
plt.show()
```

```python
plt.bar(x, Cs_1cm+hintergrund, width=1)
plt.bar(x, Cs_1cm, width=1)

plt.show()
plt.bar(x, hintergrund, width=1)
plt.show()
```

```python
plt.plot([1,2,3,4,5],[sum(Cs_1cm),sum(Cs_2cm),sum(Cs_3cm),sum(Cs_4cm),sum(Cs_5cm)],"+")
def dist(d,n,d_1):
    return n/(d+d_1)


NTC_Cont = kafe2.XYContainer([1,2,3,4,5],[sum(Cs_1cm),sum(Cs_2cm),sum(Cs_3cm),sum(Cs_4cm),sum(Cs_5cm)])
NTC_Cont.add_error(axis='x',err_val=0.1)
NTC_Cont.add_error(axis='y',err_val=2)

NTC_Fit = kafe2.Fit(NTC_Cont,dist)
NTC_Fit.do_fit()

NTC_Plot = kafe2.Plot(NTC_Fit)
NTC_Plot.plot()
plt.show()
```
