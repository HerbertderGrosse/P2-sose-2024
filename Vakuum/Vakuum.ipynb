Für die Auswertung der Messung wurde ein Video des Displays aufgenommen und mithilfe von Tesseract ausgewertet. Dabei wurden für jeden Frame alle drei Anzeigewerte ausglesen, sodass wir für jede Sekunde 30 Messwerte erhalten. 

Beim Ablesen durch Tesseract sind unvermeidbare Fehler aufgetaucht, welche sich aber auf unter 10% der Aufgenommen Messwerte beschränkt haben. Es stellte eine Herausforderung dar, diese fehlerhaft bestimmten Werte zu identifizieren und entsprechend aus dem Array zu löschen. Dadurch, dass die Hohe Framerate des Videos von 30FPS weit über derjenigen des Displays liegt, exisiteren jedoch für jeden Displaywert mehrere Bilder und entsprechend konnte der Fehler dadurch korrigiert/interpoliert werden.

Dadurch ergaben sich jeweils 10.000 Messwerte für 2.1 und 2.2 und 20.000 Messwerte für 2.3.

```python
saug = pd.read_csv('doc/2.1.csv')

def error(x):
    if np.any(x<100):
        return x*0.05
    if np.any(x> 500):
        return x*0.25
    if np.any(100<=x<=500):
        return x*(x/100)*0.05
        

saug['deltaT1'] = saug['T1'].apply(error)

plt.errorbar(saug['time[s]'],saug['T1'],yerr=saug['deltaT1'],fmt='s',lw=0.01,markersize = 0.1,label='T2')



#plt.errorbar(saug['time[s]'],saug['T1'],yerr=saug['deltaT1'],fmt='s',color='black',lw=0.01,markersize = 0.01)

plt.gcf().set_dpi(400)
plt.yscale('log')
plt.ylabel('Druck [mbar]')
plt.xlabel('Zeit [s]')
plt.legend()
plt.show()

saug = saug[907:]
saug['time[s]'] -= list(saug['time[s]'])[0]+0.0001

```

Als Fehler für die Messung wurden die Angaben aus dem Datenblatt verwendet:

- $10^{-3}\lesssim p \lesssim 10^{2}\ \mathrm{mbar}$: $\pm5\%$

- $10^{2}\lesssim p \lesssim 10^{3}\ \mathrm{mbar}$: $\pm25\%$

Wobei der zwischenbereich wie folgt modelliert wurde:

- $100\mathrm{mbar} <p< 500\mathrm{mbar}$:  $\pm \frac{p}{500\,\mathrm{mbar}}\%$




$$
dS = \frac{dV}{dt}
$$

Aus der idealen Gasgleichung

$$
pV = nRT
$$

folgt:
$$
V(t) = V_{0}\frac{p_{0}}{p(t)}
$$

und somit:
$$
dS = -V_{0}\,\frac{p_{0}}{\dot{p}^2(t)}\,p(t)
$$
Integration liefert:
$$
S(t) = -\frac{V}{t-t_{0}}\ln\left(\frac{p}{p_{0}}\right)
$$

```python
def S(t,p,V=10.5,p0=1000):
    return -V/t*np.log(p/p0)
saug['S'] = S(saug['time[s]'],saug['T1'])
result = saug.groupby('T1')['S'].agg(['mean', 'std'])

plt.errorbar(result.index,result['mean'],yerr=result['std'],xerr=error(result.index),fmt='s',color='black',lw=0.5,markersize = 0.5)
plt.xlabel('Druck [mbar]')
plt.ylabel(r'Saugvermögen [$\frac{l}{s}$]')
plt.xscale('log')

#Bereiche Definineren
plt.fill_betweenx((0,4),700,1000,alpha=0.2,label='Bereich I')
plt.fill_betweenx((0,4),200,700,alpha=0.2,label='Bereich II')
plt.fill_betweenx((0,4),0.05,200,alpha=0.2,label='Bereich III')

plt.legend()

plt.gcf().set_dpi(400)
```

Bei hohen Drücken(Bereich III) seigt die Saugleistung immer weiter an, da sich die Pumpeffozienz immer weiter erhöht. Desweiteren ist zu erkennen, dass Bereich II in auf der Logarithmischen Skala recht klein erscheint, alos das die maximale Saugleistung nur in einem relativ kleinen Druckbereich erziehlt wird. Bei weiter fallendem Druck nimmt die Saugleistung immer weiter ab (Bereich III).

Bereich 2 wird an das Modell

$$
p(t) = p_{0}\,e^{-c\,t}
$$
mit
$$
c = \frac{S}{V}
$$
angepasst.

```python
V = 10.5
def exp(t,S,t0):
    return 1000*np.exp(-S/V*(t-t0))

inter = saug[50:200]

res2 = inter.groupby('T1')['time[s]'].agg(['mean', 'std'])

fit = kafe2.xy_fit(exp,res2['mean'],res2.index,y_error=error(res2.index))
kafe2.plot()
#p = kafe2.Plot([fit])
#saug['S'][1600:2200].mean()
```

Der Fit scheint mit einer $\chi^2$-Wahrscheinlihckeit von 1 sehr gut zu sein.
Wie zu erwarten liegt der hierbei bestimmte wert für das Saugvermögen nahe bei dem, was man auf dem Graphaen in Bereich II ablesen kann.

$

### Aufgabe 3.1: Statische Kalibration von T3

Führen Sie mit Hilfe des Referenzvolumes RV eine statische Kalibration des Vakuummeters T3 durch.

---

Bei der Kalibration wird immer wieder das Volumen RV auf Atmosphärendruck gebracht und anschließend mit der Glocke verbunden. Dabei wird nach jeder Iteration der Druck in der Glocke aufgenommen bis $\sim 80\,mBar$ erreicht sind. im folgenden kann daraus der Expansionskoeffizient $\epsilon$ der Anordnung bestimmt werden. 


Als Modell für den Fit verwenden wir $\Delta p =\frac {V_{\rm{RV}} }{ V_{\rm{RZ}} }p_{\rm{atm}} = \epsilon \,  p_{\rm{atm}} \Rightarrow p_n= p_0 +\frac{n}{\epsilon} \,  p_{\rm{atm}}$ weil so in der Vorbereitung angegeben. Als Fehler wird $8\,\%$ auf die Druckmessung verwendet, da angegeben ist dass das als Abweichung von der Linearität zu erwarten ist. 

```python
stat=np.array([0,3.58,7.32,11.2,15.1,17,23.4,28.0,32.1,37.4,42.0,47.0,52.2,60.8,66.0,71.3,77.6,85.6])
 
del_stat=stat[1:]-np.roll(stat,1)[1:]

def iterations(n,epsilon=300,patm=1000,p0=0):
    return p0+n*patm*epsilon
Kali_Cont = kafe2.XYContainer(range(len(stat)), stat)
Kali_Cont.add_error(axis='x',err_val=0.001)
Kali_Cont.add_error(axis='y',err_val=0.08, relative=True)
Kali_Cont.axis_labels=['Anzahl Iterationen [n]','Druck [mBar]']

Kali_Fit = kafe2.Fit(Kali_Cont,iterations)
Kali_Fit.add_parameter_constraint("patm", 1000,1)
Kali_Fit.fix_parameter("p0", 0)

Kali_Fit.do_fit()

Kali_Plot = kafe2.Plot(Kali_Fit)
Kali_Plot.plot()

plt.show()

Kali_Fit.parameter_values[0]
```

Mit $V_{\rm{RZ}}​=10.0\,l, \ V_{\rm{RV​}}=0.034 l $ erwarten wir für $\epsilon =  \frac {V_{\rm{RV}} }{ V_{\rm{RZ}} } = 0.0034$ aus dem Fit erhalten wir $\epsilon=0.004074\pm0.00008$. Zu sehn ist dass die Daten nicht linear sind, was den Erwartungen entspricht. Für hohe Drücke weichen die Daten wie erwartet ab. Das $\frac{\chi^2}{ndf}=1.961$ ist sehr hoch und die Wahrscheinlichkeit mit $0.0102$ recht gering. Die Daten entsprechen somit nicht sonderlich dem Modell was die Abweichung erklärt.

```python



durchschlag = np.array([[1000,4225],[500,1223],[249,1100],[122,867],[62.4,682],[31.8,525],[15.1,407],[7.52,387],[3.52,323],[1.72,300],[0.842,399],[0.433,443],[0.200,425],[0.0833,525]]) #[Druck[mbar],Spannung]

durchschlagTMP = np.array([[0.02,1000],[0.025,910],[0.03,835],[0.35,777],[0.46,710],[0.053,601],[0.056,585],[0.06,576],[0.065,566],[0.07,559],[0.075,553],[0.08,545],[0.083,542],[0.086,539],[0.09,537],[0.095,532],[0.1,530],[0.11,525]])
durchschlagTMP_p_err = durchschlagTMP[:,0]*0.3
durchschlag_p_err = durchschlag[:,0]*0.1
plt.errorbar(durchschlag[:,0],durchschlag[:,1],xerr=durchschlag_p_err,yerr=100, label="Messung mit DSP",fmt=".")
plt.errorbar(durchschlagTMP[:,0],durchschlagTMP[:,1],xerr=durchschlagTMP_p_err,yerr=100,label="Messung mit TMP",fmt=".")
plt.xscale('log')
plt.xlabel("Druck [mBar]")
plt.ylabel("Durchschlagsspannung [V]")
plt.legend()
plt.title("Durchschlagsspannung über den Druck aufgetragen (Druck logarithmisch)")
def iterations(x,b=1.7,c=380,a=25):
    return a*((np.log(x)-b))**2+c

Durchschlag_Cont = kafe2.XYContainer(np.append(durchschlag[:,0],durchschlagTMP[:,0])[1:],np.append(durchschlag[:,1],durchschlagTMP[:,1])[1:])
Durchschlag_Cont.add_error(axis='x',err_val=0.05,relative=True)
Durchschlag_Cont.add_error(axis='y',err_val=100)
Durchschlag_Cont.axis_labels=["Druck [mBar]","Durchschlagsspannung [V]"]

Durchschlag_Fit = kafe2.Fit(Durchschlag_Cont,iterations)
Durchschlag_Fit.do_fit()

Durchschlag_Plot = kafe2.Plot(Durchschlag_Fit)
Durchschlag_Plot.x_scale = "log"
Durchschlag_Plot.plot()


plt.show()


p=u.ufloat(np.exp(Durchschlag_Fit.parameter_values[0]), np.exp(Durchschlag_Fit.parameter_errors[0]))


def lambd(p_rz,k=1.38064852*10**-20,T=298.15,M=28.96,N=6.02214086*10**23,rho=807,f=0.74):
    return (k*T)/(np.pi*p_rz*( (3*M*f)/(4*np.pi*N*rho) )**(2/3))
print(lambd(p*100)),np.exp(Durchschlag_Fit.parameter_values[0])
```

Bei der Messung mit der TMP wurde das Ionisations-Vakuummeter zu lange abgelesen. Dadurch gibt es zwei Messpunkte die  aus der Reihe fallen. Zu Begin der Messung mit der TMP  war der Druck so niedrig, dass die Durchschlagsspannung mit der Spannungsquelle nicht erreichbar war. Nachdem der Druck fiel war erst sehr spät eine neblige Erscheinung zu sehen. Deswegen wurde keine Durchschlagsspannung über $1000\,v$ gemessen. Vermutlich ist hier ein Fehler in der Versuchsdurchführung verantwortlich. 

Zunächst tritt ein Blitzartiger gepulster Blitz auf was übergeht in ein Glimmen zwischen den Kugeln. Teilweise war das Glimmen nicht sichtbar, deswegen wurde die maximale Spannung am Multimeter verwendet. Bei geringem Druck füllte das Glimmen einen großen Teil der Kuppel aus.

Die Durchschlagsspannung sinkt zunächst, da durch die größer werdende mittlere freie Weglänge $\lambda$ die Elektronen größere Geschwindigkeiten erreichen und deswegen leichter ionisieren können. 
Das Minimum der Durchschlagsspannung liegt bei ca. $0.161\,mBar$. Das entspricht dem Druck bei dem $\lambda$ dem Abstand der Kugelelektroden entspricht.

Demnach ergibt sich für den Abstand der Kugelelektroden $\lambda = \frac{k_{\rm{B}} T}{\pi \left(\sqrt[3]{\frac{3 M_m f}{4\pi N_A \rho_{\rm{fl}}}}\right)^2 p_{\rm{RZ}} }=1.7\pm0.12\,mm$

Danach steigt die Durchschlagsspannung wieder an da die Wahrscheinlichkeit sinkt einTeilchen zu treffen und deswegen das Gas nicht so leicht ionisiert wird.
Der Druck lässt sich bei kleinen Drücken nichtmehr gut einstellen weil durch Lecks Luft eintritt und die Zeitdauer bis sich ein konstanter Druck einstellt wenn das Ventil teilweise verschlossen ist steigt. 


### Aufgabe 3.3: Aufdampfen von Indium

Dampfen Sie mit Hilfe einer Schablone bei zwei verschiedenen Drucken Indium auf eine Plexiglasplatte auf.

---

In diesem Versuch wird Indium bei zwei verschiedenen Drücken auf eine Plexiglasscheibe gedampft 

Das Indium befindet sich in einem Schiffchen aus dünnen Blech, welches durch einen Strom beheizt werden kann. Darüber befindet sich ein Blech mit einem kleinen Loch durch das das verdampfte Indium austreten kann. Das Indium wird dadurch auf eine schwenkbare Plexiglasscheibe aufgedampft. 

Nachdem die Erste Durchführung misslang, da das Schiffchen mit dem Indium durchgebrannt war, wurde die Kammer auf $1.2\cdot10^{-4}\,mBar$ evakuiert. Anschließend wurde das Indium mit etwa $40\,A$ erwärmt. Das aufgedampfte Indium war leicht auszumachen und  hatte eine relativ scharfe Grenze. Nun wurden die Pumpen abgeschaltet und der DrucK stieg an. Jetzt wurde das Indium wider erhitzt. Dazu wurde mehr Strom (ca. $45\,A$) für eine längere Zeit benötigt, da die Luft Wärme abführte. Es dauerte länger bis sich ein sichtbarer Fleck bildete. Der Fleck war weniger scharf begrenzt und weniger intensiv. Das liegt daran das das Indium an den Molekülen der Luft stärker gestreut wird.
